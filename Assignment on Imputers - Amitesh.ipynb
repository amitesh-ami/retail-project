# **Loading the dataset**

```python
import pandas as pd 
import numpy as np
import time
```

```python
df = pd.read_csv('Retail-data.csv')
```

```python
df.head(10)
```

```python
# converting the time object into separate columns 

df['hour'] = pd.to_datetime(df['hour'])
df['year'] = df['hour'].dt.year
df['month'] = df['hour'].dt.month
df['week'] = df['hour'].dt.week
df['day'] = df['hour'].dt.day
df['name'] = df['hour'].dt.weekday
```

```python
df.describe().transpose()
```

# **Imputation techniques**

```python
from sklearn.impute import SimpleImputer
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
#strategy= mean,median,mode    : mean is default
#verbose=0 for column , 1 for row    : 0 is default

si_mf = SimpleImputer(strategy='most_frequent',add_indicator=True)
```

```python
df['order_name'].isnull().sum()
```

```python
#dealing with categorical variables
start_time = time.time()

df['order_name']=si_mf.fit_transform(df[['order_name']])
df['product_type']=si_mf.fit_transform(df[['product_type']])
df['product_title']=si_mf.fit_transform(df[['product_title']])
df['product_vendor']=si_mf.fit_transform(df[['product_vendor']])
df['variant_title']=si_mf.fit_transform(df[['variant_title']])
df['api_client_title']=si_mf.fit_transform(df[['api_client_title']])

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
df.isnull().sum()
```

```python
# dealing with numeric variables

si_mean = SimpleImputer(missing_values=0.0,strategy='mean',add_indicator=True)
```

```python

start_time = time.time()

df['order_id']=si_mean.fit_transform(df[['order_id']])
#df['hour']=si_mean.fit_transform(df[['hour']])
df['customer_id']=si_mean.fit_transform(df[['customer_id']])
df['product_id']=si_mean.fit_transform(df[['product_id']])
df['product_price']=si_mean.fit_transform(df[['product_price']])
df['variant_id']=si_mean.fit_transform(df[['variant_id']])
df['gross_sales']=si_mean.fit_transform(df[['gross_sales']])
df['discounts']=si_mean.fit_transform(df[['discounts']])
df['returns']=si_mean.fit_transform(df[['returns']])
df['net_sales']=si_mean.fit_transform(df[['net_sales']])
df['shipping']=si_mean.fit_transform(df[['shipping']])
df['taxes']=si_mean.fit_transform(df[['taxes']])
df['total_sales']=si_mean.fit_transform(df[['total_sales']])
df['ordered_item_quantity']=si_mean.fit_transform(df[['ordered_item_quantity']])
df['net_quantity']=si_mean.fit_transform(df[['net_quantity']])
df['returned_item_quantity']=si_mean.fit_transform(df[['returned_item_quantity']])

print("--- %s seconds ---" % (time.time() - start_time))
```
